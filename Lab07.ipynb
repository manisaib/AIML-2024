```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
cancer['feature_names']
```

```python
a=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
a.info()
```

```python
a.describe()
```

```python
np.sum(pd.isnull(a.sum()))
```

```python
cancer['target']
```

```python
a['cancer']=pd.DataFrame(cancer['target'])
a.head()
```

```python
a_feat=a.drop('cancer',axis=1)
a_feat.head()
```

```python
a_target = a['cancer']
a_target.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(a_feat,a_target,test_size=0.30,random_state=101)
```

```python
y_train.head()
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(X_train,y_train)
```

```python
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
"prediction rate:",round(np.mean(predictions!=y_test),3)
```
